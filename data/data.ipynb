*Calculating the top 10 spenders percentage of the total independent expenditure of that year*

```python
# isolates the dataframe to only spenders in 2024
yr2024 = frame.loc[frame['report_year'] == 2024]

# grouping committees based on their total spending within in election
grouped_committees = yr2024.groupby(['committee_name', 'candidate_party'])['expenditure_amount'].sum().reset_index()

# now I have to find the top 15 spenders of that year, and their name and party
top_15_spenders = grouped_committees.sort_values(by='expenditure_amount', ascending=False).head(15)
```

```python
# finding total expenditure of 2024
total = yr2024['expenditure_amount'].sum()
total
```

```python
# now to find percentages of top 10 pacs in the total amount of expnediture in 2024
top_15_spenders['percentage_exp'] = ((top_15_spenders['expenditure_amount'] /total) * 100).round(2)
top_15_spenders
```

```python
# Calculating the other
total_percentage = top_15_spenders['percentage_exp'].sum()

print(f'Remaining percentage after 15: {round(100-total_percentage, 2)}')
other_row = {'committee_name': 'Other PACs', 'percentage_exp': round(100-total_percentage, 2)}
top_15_spenders = pd.concat([top_15_spenders, pd.DataFrame([other_row])], ignore_index=True)

top_15_spenders
```

```python
top_15_spenders.to_json('top_15_percentages.json', orient='records')
```
